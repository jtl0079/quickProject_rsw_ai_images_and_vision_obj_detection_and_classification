```python
import os
# 导入你决定保留的、极具辨识度的流水线模块
import rsw_ai.pipeline.create_circle_image_and_create_metadata_and_write_image_metadata_and_write_circle_metadata_and_create_metadata_file as pipeline

# 确保输出目录存在
os.makedirs("../outputs/images", exist_ok=True)
os.makedirs("../outputs/metadata", exist_ok=True)

# ==========================================
# 1. 状态定义 (Status Definition)
# ==========================================
# 依照你的规划：
# width, height, radius, colors, thickness -> STATUS: fixed
# center (坐标)                            -> STATUS: algo

# [fixed] 基础固定参数
WIDTH = 300
HEIGHT = 300
RADIUS = 50
CIRCLE_COLOR = (0, 0, 0)
BACKGROUND_COLOR = (255, 255, 255)
THICKNESS = 5

# ==========================================
# 2. 批量生成逻辑 (Loop with 'algo' Center)
# ==========================================
# 这里使用一个简单的对角线移动算法 (algo) 作为示范，生成 5 张图片进行 test a
# 你可以根据实际的 test a 需求修改这个循环和算法
num_images = 5

print(f"开始生成测试数据集 (共 {num_images} 张)...")

for i in range(num_images):
    # [algo] 算法逻辑坐标：让圆心随着循环从 (50, 50) 对角线移动到 (250, 250)
    # 算法公式示例：center_x = 50 + i * 50, center_y = 50 + i * 50
    step = i * 50
    algo_center = (50 + step, 50 + step)
    
    # 为了防止文件覆盖，将索引 i 动态编入文件名
    img_path = f"../outputs/images/test_a_circle_{i}.png"
    json_path = f"../outputs/metadata/test_a_circle_{i}.json"
    
    # 调用保留的完整流水线函数
    pipeline.create_circle_image_and_create_metadata_and_write_image_metadata_and_write_circle_metadata_and_create_metadata_file(
        filename=img_path,
        json_filename=json_path,
        width=WIDTH,
        height=HEIGHT,
        radius=RADIUS,
        circle_color=CIRCLE_COLOR,
        background_color=BACKGROUND_COLOR,
        center=algo_center,       # 传入算法动态计算的坐标
        thickness=THICKNESS,
    )
    
    print(f" 已生成第 {i+1}/{num_images} 张: 坐标为 {algo_center}")

print("\n[STATUS] 所有测试数据生成完毕！可以开始进行 test a。")
```
