```python
from rsw_ai.model.DatasetGenerator import DatasetGenerator
import random
generator = DatasetGenerator()

generator.create_circle_dataset(
    image_filename="../outputs/images/circle.png",
    metadata_filename="../outputs/metadata/circle.json",
    count= 10,
    center= lambda param_a:(
        random.randint(0,499),
        random.randint(0,499)
    ),
    radius= 50,
    thickness=10,
)

```

```python
from rsw_ai.backend.read_metadata_file import read_metadata_file
from rsw_ai.backend.metadata_to_yolo_txt import metadata_to_yolo_txt
from rsw_ai.backend.write_text_file import write_text_file

metadata = read_metadata_file("../outputs/metadata/dummy_circle_api.json")

yolo_txt_data = metadata_to_yolo_txt(metadata=metadata)

write_text_file(
    output_file="qwe.txt",
    lines=yolo_txt_data,
)

print(metadata, "\n", yolo_txt_data)
```
